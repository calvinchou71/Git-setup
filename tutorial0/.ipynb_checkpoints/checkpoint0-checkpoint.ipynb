```python
# optional challenge - fizzbuzz
# you will need to use both iteration and control flow 
# go through all numbers from 1 to 100 in order
# if the number is a multiple of 3, print fizz
# if the number is a multiple of 5, print buzz
# if the number is a multiple of 3 and 5, print fizzbuzz and NOTHING ELSE
# if the number is neither a multiple of 3 nor a multiple of 5, print the number
list1 = list(range(1,101))
for n in list1:
    if n%3==0 and n%5==0:
        print("fizzbuzz")
    elif n%3==0:
        print("fizz")
    elif n%5==0:
        print("buzz")
    else:
        print(n)
```

```python
# load in the "starbucks.csv" dataset
# refer to how we read the cereal.csv dataset in the tutorial
import pandas as pd
starbucks = pd.read_csv("../data/starbucks.csv")
starbucks.head()
```

```python
# output the calories, sugars, and protein columns only of every 40th row. 
starbucks.iloc[::40][["calories", "sugars", "protein"]]
```

```python
# select all rows with more than and including 400 calories
starbucks.loc[starbucks["calories"]>=400]
```

```python
# select all rows whose vitamin c content is higher than the iron content
starbucks.loc[starbucks["vitamin c"] > starbucks["iron"]]
```

```python
# create a new column containing the caffeine per calories of each drink
starbucks["caffeine per calories"] = starbucks["caffeine"]/starbucks["calories"]
starbucks.head()
```

```python
# what is the average calorie across all items?
starbucks["calories"].mean()
```

```python
# how many different categories of beverages are there?
starbucks["beverage_category"].nunique()
```

```python
# what is the average # calories for each beverage category?
starbucks.groupby("beverage_category")["calories"].mean()
```

```python
# plot the distribution of the number of calories in drinks with a histogram
starbucks["calories"].plot.hist(title = "calories")
```

```python
# plot calories against total fat with a scatterplot
starbucks.plot.scatter(x="total fat", y="calories")
```
